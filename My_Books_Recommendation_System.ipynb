```python
df['description'] = df['description'].str.split()
df['description'] = df['description'].apply(lambda x: [i.replace(" ", "") for i in x])
df['description'] = df['description'].apply(lambda x: " ".join(x))
df
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# importing stopwords
from nltk.corpus import stopwords
stopwords.words('english')

# importing regular expressions
import re

# importing punctuations
import string
from string import punctuation

#importing Porter stemmer

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer() 
```

```python
def desc_transform(desc):
    desc = desc.lower() #Lowercase
    
    # Removing special characters and numbers
    desc = re.sub('[^a-zA-Z \n\.]', '', desc)
    
    # separating sentences in word
    desc = nltk.word_tokenize(desc)
    
    # Removing stopwords and punctuations
    lst = []
    for i in desc:
        if i not in stopwords.words('english') and i not in string.punctuation:
            lst.append(i)
            
    #stemming
    desc = lst[:]
    lst.clear()
    for i in desc:
        lst.append(ps.stem(i))
    
    desc = " ".join(lst)
    lst.clear()
    
    return desc
```

```python
df['desc'] = df['description'].apply(desc_transform)
```

```python
new_df = df[['title','desc','image_url']]
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['desc']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
similarity.shape
```

```python
# Similarity of first book with all books
similarity[0]
```

```python
#sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])[1:6]
```

```python
def recommend(book):
    book_index = new_df[new_df['title'] == book].index[0]
    distances = similarity[book_index]
    book_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]
    
    for i in book_list:
        print(new_df.iloc[i[0]].title)
        print(new_df.iloc[i[0]].image_url)
```

```python
recommend('Walk Two Moons')
```
